## STACK
Stack adalah struktur data linier yang mengikuti prinsip Last-In/First-Out(LIFO) yang berarti elemen terakhir yang ditambahkan adalah elemen pertama yang dihapus. Dalam sebuah stack, penyisipan dan penghapusan terjadi di ujung yang sama, yang disebut puncak stack.

#### -> Menggunakan List
- Metode append() untuk menambahkan elemen ke akhir daftar.
- Metode pop() untuk menghapus dan mengembalikan elemen terakhir dari daftar.

#### -> Menggunakan collections.deque
- Metode append() untuk menambahkan elemen ke ujung kanan deque
- Metode pop() untuk menghapus dan mengembalikan elemen dari ujung kanan

#### -> Stack itu cuma 5 operasi utama:
1. Push → tambah
2. Pop → hapus atas
3. Peek → lihat atas
4. isEmpty → cek kosong
5. Size → hitung jumlah

```python
# VERSI MENGGUNAKAN CLASS

class stack:
    def __init__(nama):
        nama.stack = []
    # push
    def push(nama, elemen):
        nama.stack.append(elemen)
    # pop
    def pop(nama):
        if nama.isEmpty():
            return "Stack is empty"
        return nama.stack.pop()
    # peek
    def peek(nama):
        if nama.isEmpty():
          return "Stack is empty"
        return nama.stack[-1]
    # isEmpty
    def isEmpty(nama):
        return len(nama.stack) == 0
    # size
    def size(nama):
        return len(nama.stack)


myStack = stack()

myStack.push('Hilda')
myStack.push('Mahligatul')
myStack.push('Uyun')

print("Stack: ", myStack.stack)
print("Pop: ", myStack.pop())
print("Stack setelah Pop: ", myStack.stack)
print("Elemen teratas: ", myStack.peek())
print("isEmpty: ", myStack.isEmpty())
print("Jumlah elemen: ", myStack.size())
```

## Ekspresi Aritmik Infix, Prefix, Postfix

### Infix
Ekspresi infiks adalah ekspresi matematika di mana operator ditempatkan di antara operandnya . Ini adalah notasi matematika paling umum yang digunakan oleh manusia. Misalnya, ekspresi "2 + 3" adalah ekspresi infiks, di mana operator "+" ditempatkan di antara operand "2" dan "3".

#### Cara umum penulisan ekspresi infiks:
- Notasi infiks adalah notasi yang paling kita kenal. Misalnya, ekspresi "2 + 3" ditulis dalam notasi infiks.
- Dalam notasi infiks, operator ditempatkan di antara operan yang dioperasikannya. Misalnya, dalam ekspresi "2 + 3", operator penjumlahan "+" ditempatkan di antara operan "2" dan "3".
- Tanda kurung digunakan dalam notasi infiks untuk menentukan urutan operasi yang harus dilakukan. Misalnya, dalam ekspresi "(2 + 3) * 4", tanda kurung menunjukkan bahwa operasi penjumlahan harus dilakukan sebelum operasi perkalian.

#### Aturan prioritas operator
- Tanda kurung ()	Paling tinggi
- Eksponen ^	Tinggi
- Perkalian *	Sedang
- Divisi /	Sedang
- Penambahan +	Rendah
- Pengurangan -	Rendah

#### Contoh:
A + B

5 - 3

X * Y

8 / 2

2 + 3 * 4

(1 + 2) * 3

### Postfix
Ekspresi postfix, juga dikenal sebagai Notasi Polandia Terbalik (RPN) , adalah notasi matematika di mana operator mengikuti operandnya . Ini berbeda dari notasi infix yang lebih umum, di mana operator ditempatkan di antara operandnya.

Dalam notasi postfix, operan ditulis terlebih dahulu, diikuti oleh operator. Misalnya, ekspresi infix "5 + 2" akan ditulis sebagai "5 2 +" dalam notasi postfix.

#### Aturan Dasar Konversi Infix → Postfix

Biasanya pakai stack (tumpukan).

Prioritas operator:

'^' (pangkat)

'*' dan /
  
'+' dan -

#### Langkah-langkahnya:
Kalau operand (A, B, angka) → langsung masuk ke output

Kalau operator → cek prioritas:

lebih tinggi → masuk stack

lebih rendah → keluarkan isi stack dulu

Kalau ( → masuk stack

Kalau ) → keluarkan sampai ketemu (

#### Contoh:
Infix: A + B → Postfix: A B +

Infix: A * B → Postfix: A B *

Infix: A + B * C → Postfix: A B C * +

```python
def infix_to_postfix(expression):
    # prioritas operator
    precedence = {
        '^': 3,
        '*': 2,
        '/': 2,
        '+': 1,
        '-': 1
    }

    stack = []      # untuk operator
    output = []     # hasil postfix

    for char in expression:
        # kalau operand (huruf/angka)
        if char.isalnum():
            output.append(char)

        # kalau buka kurung
        elif char == '(':
            stack.append(char)

        # kalau tutup kurung
        elif char == ')':
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            stack.pop()  # buang '('

        # kalau operator
        else:
            while (stack and stack[-1] != '(' and
                   precedence.get(char, 0) <= precedence.get(stack[-1], 0)):
                output.append(stack.pop())
            stack.append(char)

    # keluarkan sisa stack
    while stack:
        output.append(stack.pop())

    return ''.join(output)


# contoh pemakaian
exp = "A+B*C"
print("Infix  :", exp)
print("Postfix:", infix_to_postfix(exp))
```

### Prefix
Prefix (Polish Notation) adalah penulisan ekspresi di mana operator ditaruh di depan operand.

#### Contoh dasar:
Infix: A + B → Prefix: + A B

Infix: A * B → Prefix: * A B

Infix: A + B * C → Prefix: + A * B C

#### Cara Konversi Infix → Prefix (cara gampang)

*Langkah:*
Balik ekspresi infix

Tukar ( jadi ) dan sebaliknya

Ubah ke postfix

Balik lagi hasilnya → jadi prefix



```python
def infix_to_prefix(expression):
    # fungsi prioritas operator
    def precedence(op):
        if op == '+' or op == '-':
            return 1
        elif op == '*' or op == '/':
            return 2
        elif op == '^':
            return 3
        return 0

    # fungsi konversi infix ke postfix
    def infix_to_postfix(exp):
        stack = []
        output = []

        for char in exp:
            if char.isalnum():
                output.append(char)
            elif char == '(':
                stack.append(char)
            elif char == ')':
                while stack and stack[-1] != '(':
                    output.append(stack.pop())
                stack.pop()
            else:
                while (stack and precedence(char) <= precedence(stack[-1])):
                    output.append(stack.pop())
                stack.append(char)

        while stack:
            output.append(stack.pop())

        return ''.join(output)

    # STEP 1: balik ekspresi
    expression = expression[::-1]

    # STEP 2: tukar kurung
    new_exp = ""
    for char in expression:
        if char == '(':
            new_exp += ')'
        elif char == ')':
            new_exp += '('
        else:
            new_exp += char

    # STEP 3: infix -> postfix
    postfix = infix_to_postfix(new_exp)

    # STEP 4: balik lagi hasilnya
    prefix = postfix[::-1]

    return prefix


# contoh pemakaian
exp = "A+B*C"
print("Infix  :", exp)
print("Prefix :", infix_to_prefix(exp))
```

#### 1. Buble Sort
Bandingin dua elemen yang bersebelahan, kalau salah urut → tukar.

```python
data = [5, 3, 8, 4, 2]

n = len(data)

for i in range(n):
    for j in range(0, n-i-1):
        if data[j] > data[j+1]:
            # tukar
            data[j], data[j+1] = data[j+1], data[j]

print("Hasil:", data)
```

```python
def bubble_sort(data):
    n = len(data)
    
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                # tukar posisi
                data[j], data[j + 1] = data[j + 1], data[j]
    
    return data


# contoh penggunaan
angka = [5, 3, 8, 4, 2]
hasil = bubble_sort(angka)

print("Hasil sorting:", hasil)
```

### Selection Sort
Cari nilai terkecil, lalu tukar ke depan, Jadi tiap langkah langsung “milih” yang paling kecil.

```python
lst = [4,7,1,9,3,10,6]

print(lst)
for i in range(0,len(lst)-1):
    min=i
    for j in range(i+1,len(lst)):
        if lst[j]<lst[min]:
            min = j
    lst[i],lst[min]=lst[min],lst[i]

print(lst)
```

### 3.Insertion Sort
Ambil satu data, lalu sisipkan ke posisi yang benar, Kayak nyusun kartu di tangan 🃏

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]  # yang mau disisipkan
        j = i - 1

        # geser elemen yang lebih besar
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        # masukkan key ke posisi yang tepat
        arr[j + 1] = key

    return arr

data = [5, 2, 4, 1]
print(insertion_sort(data))
```

# Queue (Antrian)

## Pengertian Queue
Queue (antrian) adalah struktur data yang menggunakan prinsip FIFO (First In First Out), yaitu elemen yang pertama masuk akan menjadi elemen pertama yang keluar.

Contoh penerapan:

Antrian di kasir

Antrian pelayanan

## Operasi Dasar queue
Operasi yang umum digunakan pada queue:

1. Enqueue: menambahkan elemen ke bagian belakang
2. Dequeue: menghapus elemen dari bagian depan
3. Front/Peek: melihat elemen paling depan tanpa menghapus
4. isEmpty: mengecek apakah queue kosong

## Implementasi Queue Menggunakan List
queue = []

### enqueue
queue.append("A")

queue.append("B")

queue.append("C")

print(queue)  # ['A', 'B', 'C']

### dequeue
queue.pop(0)

print(queue)  # ['B', 'C']

*Kelemahan:*

Operasi pop(0) kurang efisien karena harus menggeser seluruh elemen.

## Implementasi Queue Menggunakan deque

queue = deque()

### enqueue
queue.append("A")

queue.append("B")

### dequeue
queue.popleft()

print(queue)

*Kelebihan:*

Lebih efisien dibandingkan list karena tidak perlu menggeser elemen.

## Deque (Double-Ended Queue)
Pengertian Deque

Deque adalah struktur data yang memungkinkan penambahan dan penghapusan elemen dari kedua sisi, yaitu depan dan belakang.

## Operasi pada Deque
1. append(): menambahkan elemen ke belakang
2. appendleft(): menambahkan elemen ke depan
3. pop(): menghapus elemen dari belakang
4. popleft(): menghapus elemen dari depan

## Contoh Penggunaan Deque

```python
from collections import deque

d = deque()

d.append(1)
d.append(2)
d.appendleft(0)

print(d)  # deque([0, 1, 2])

d.pop()
print(d)  # deque([0, 1])

d.popleft()
print(d)  # deque([1])
```

## Perbedaan Queue dan Deque

### Queue:

Menggunakan prinsip FIFO

Penambahan di belakang, penghapusan di depan

Lebih sederhana

### Deque:

Penambahan dan penghapusan bisa di dua sisi

Lebih fleksibel

Dapat digunakan sebagai queue maupun stack

## Searching (Pencarian Data)

Searching adalah proses mencari suatu data (nilai/elemen) di dalam kumpulan data (array, list, atau struktur data lainnya) untuk mengetahui apakah data tersebut ada dan di posisi mana.

### Pembagian Searching

#### 1. Sequential Search (Linear Search)
Mencari data satu per satu dari awal sampai akhir

- Ciri-ciri:
- 
Tidak perlu data terurut

Sederhana

Lebih lambat untuk data besar

```python
def linear_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

data = [10, 20, 30, 40]
print(linear_search(data, 30))
```

#### 2. Binary Search
Mencari dengan cara membagi data menjadi dua bagian

- Ciri-ciri:
- 
Data harus terurut

Lebih cepat dari linear search

Menggunakan konsep “bagi dua”

```python
def binary_search(data, target):
    kiri = 0
    kanan = len(data) - 1

    while kiri <= kanan:
        tengah = (kiri + kanan) // 2

        if data[tengah] == target:
            return tengah
        elif data[tengah] < target:
            kiri = tengah + 1
        else:
            kanan = tengah - 1

    return -1

data = [10, 20, 30, 40]
print(binary_search(data, 30))
```

## Hashing

Hashing adalah teknik untuk menyimpan dan mencari data dengan cara mengubah key menjadi index menggunakan hash function. Jadi, Data langsung tahu tempatnya tanpa harus dicari satu-satu.

### Hash Function
Fungsi untuk menentukan index dari sebuah key

```python
def hash_function(key):
    return key % 10
```

### Hash Table
Tempat menyimpan data (biasanya array/list)

```python
size = 10
hash_table = [[] for _ in range(size)]
```

### Operasi pada Hash Table (CRUD)

```python
# Insert atau tambah
def insert(key):
    index = hash_function(key)
    hash_table[index].append(key)

# Search atau cari
def search(key):
    index = hash_function(key)
    return key in hash_table[index]

# Delete atau hapus
def delete(key):
    index = hash_function(key)
    if key in hash_table[index]:
        hash_table[index].remove(key)

# Update
def update(old, new):
    delete(old)
    insert(new)
```

## Merge Sort

Merge Sort adalah algoritma pengurutan yang menggunakan metode divide and conquer, yaitu membagi data menjadi bagian kecil, lalu menggabungkannya kembali dalam keadaan terurut.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        kiri = arr[:mid]
        kanan = arr[mid:]

        # Rekursif (bagi terus)
        merge_sort(kiri)
        merge_sort(kanan)

        i = j = k = 0

        # Gabung & urutkan
        while i < len(kiri) and j < len(kanan):
            if kiri[i] < kanan[j]:
                arr[k] = kiri[i]
                i += 1
            else:
                arr[k] = kanan[j]
                j += 1
            k += 1

        # Sisa kiri
        while i < len(kiri):
            arr[k] = kiri[i]
            i += 1
            k += 1

        # Sisa kanan
        while j < len(kanan):
            arr[k] = kanan[j]
            j += 1
            k += 1


# Contoh penggunaan
data = [38, 27, 43, 3]
merge_sort(data)
print(data)
```

## Quick Sort
Quick Sort adalah algoritma sorting yang menggunakan metode divide and conquer dengan memilih satu elemen sebagai pivot, lalu membagi data menjadi dua bagian.

### Cara Kerja

Misal:

[10, 7, 8, 9, 1, 5]

#### 1. Pilih Pivot

Misal ambil terakhir:

pivot = 5

#### 2. Partisi (Pisahkan)
   
kecil dari 5 → [1]

besar dari 5 → [10, 7, 8, 9]

#### 3. Gabung
   
[1] + [5] + [10, 7, 8, 9]

#### 4. Ulangi lagi (rekursif)

Sampai semua terurut:

[1, 5, 7, 8, 9, 10]

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[-1]  # ambil elemen terakhir
    kiri = []
    kanan = []

    for i in range(len(arr) - 1):
        if arr[i] < pivot:
            kiri.append(arr[i])
        else:
            kanan.append(arr[i])

    return quick_sort(kiri) + [pivot] + quick_sort(kanan)


# Contoh penggunaan
data = [10, 7, 8, 9, 1, 5]
hasil = quick_sort(data)
print(hasil)
```
